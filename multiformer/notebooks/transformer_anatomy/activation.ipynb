```python
from importlib.metadata import version
import torch

print("TORCH VERSION :", version("torch"))
device = (
    "cuda"
    if torch.cuda.is_available()
    else "mps" if torch.backend.mps.is_available() else "cpu"
)
print("GPU  : ", device)
```

```python
import torch.nn as nn
```

```python
x = torch.linspace(-10, 10, 100)
relu_out = nn.ReLU()(x)
gelu_out = nn.GELU()(x)
silu_out = nn.SiLU()(x)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=relu_out, mode="lines", name="ReLU"))
fig.add_trace(go.Scatter(x=x, y=gelu_out, mode="lines", name="GeLU"))
fig.add_trace(go.Scatter(x=x, y=silu_out, mode="lines", name="SiLU"))

fig.show()
```
