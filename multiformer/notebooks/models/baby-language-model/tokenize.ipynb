```python
# Let's calculate the total number of words to take a rough approximation of vocab size
def process_chunks(dataset, chunk_size=1000):
    unique_words = set()
    for i in range(0, len(dataset), chunk_size):
        chunk = dataset[i:i+chunk_size]['text']
        for text in chunk:
            unique_words.update(set(text.split()))
    return unique_words

# Process dataset in chunks
unique_words = process_chunks(dataset_val)

# Print total unique words
print("Total unique words separated by space:", len(unique_words))

#let's delete all the vars from this cell 
del unique_words
```

```python
vocab_size = 2000
model_prefix = f"tokenzer[{vocab_size}]"
```

```python
## The SentencePieceTrainer don't except the dataloader object input type is text by dafult.
```

```python
%time
sp.SentencePieceTrainer.train(input=dataset_val['text'],
                                   model_prefix=model_prefix,
                                   model_type="bpe",
                                   vocab_size=vocab_size,                                   self_test_sample_size=0,
                                   input_format="text",
                                   character_coverage=1.0,
                                   num_threads=os.cpu_count(),
                                   split_digits=True,
                                   allow_whitespace_only_pieces=True,
                                   byte_fallback=True,
                                   unk_surface=r" \342\201\207 ",
                                   normalization_rule_name="identity")
```
